```python
import pandas as pd
import networkx as nx
import itertools



file_path = '/content/drive/MyDrive/CEE201/pivoted_network_travel_times-Group_44.xlsx'
travel_times = pd.read_excel(file_path)

#capacity_file_path = '/content/drive/MyDrive/capacity_file-Group44.xlsx'
#capacity_values = pd.read_excel(capacity_file_path)

#file_path = '/content/drive/MyDrive/CEE201/pivoted_network_travel_times-Group_44.xlsx'
#travel_times = pd.read_csv(file_path)


values = []
for i in range(0, 76):
 values.append(travel_times.iloc[i, 1])

print(values[0])
N = nx.DiGraph()
N.add_nodes_from(range(1, 25))
N.add_edge(1, 2, weight=values[0], color='blue')
N.add_edge(1, 3, weight=values[1], color='blue')
N.add_edge(2, 1, weight=values[2], color='blue')
N.add_edge(2, 6, weight=values[3], color='blue')
N.add_edge(3, 1, weight=values[4], color='blue')
N.add_edge(3, 4, weight=values[5], color='blue')
N.add_edge(3, 12, weight=values[6], color='blue')
N.add_edge(4, 3, weight=values[7], color='blue')
N.add_edge(4, 5, weight=values[8], color='blue')
N.add_edge(4, 11, weight=values[9], color='blue')
N.add_edge(5, 4, weight=values[10], color='blue')
N.add_edge(5, 6, weight=values[11], color='blue')
N.add_edge(5, 9, weight=values[12], color='blue')
N.add_edge(6, 2, weight=values[13], color='blue')
N.add_edge(6, 5, weight=values[14], color='blue')
N.add_edge(6, 8, weight=values[15], color='blue')
N.add_edge(7, 8, weight=values[16], color='blue')
N.add_edge(7, 18, weight=values[17], color='blue')
N.add_edge(8, 6, weight=values[18], color='blue')
N.add_edge(8, 7, weight=values[19], color='blue')
N.add_edge(8, 9, weight=values[20], color='blue')
N.add_edge(8, 16, weight=values[21], color='blue')
N.add_edge(9, 5, weight=values[22], color='blue')
N.add_edge(9, 8, weight=values[23], color='blue')
N.add_edge(9, 10, weight=values[24], color='blue')
N.add_edge(10, 9, weight=values[25], color='blue')
N.add_edge(10, 11, weight=values[26], color='blue')
N.add_edge(10, 15, weight=values[27], color='blue')
N.add_edge(10, 16, weight=values[28], color='blue')
N.add_edge(10, 17, weight=values[29], color='blue')
N.add_edge(11, 4, weight=values[30], color='blue')
N.add_edge(11, 10, weight=values[31], color='blue')
N.add_edge(11, 12, weight=values[32], color='blue')
N.add_edge(11, 14, weight=values[33], color='blue')
N.add_edge(12, 3, weight=values[34], color='blue')
N.add_edge(12, 11, weight=values[35], color='blue')
N.add_edge(12, 13, weight=values[36], color='blue')
N.add_edge(13,12, weight=values[37], color='blue')
N.add_edge(13,24, weight=values[38], color='blue')
N.add_edge(14,11, weight=values[39], color='blue')
N.add_edge(14,15, weight=values[40], color='blue')
N.add_edge(14,23, weight=values[41], color='blue')
N.add_edge(15,10, weight=values[42], color='blue')
N.add_edge(15,14, weight=values[43], color='blue')
N.add_edge(15,19, weight=values[44], color='blue')
N.add_edge(15,22, weight=values[45], color='blue')
N.add_edge(16,8, weight=values[46], color='blue')
N.add_edge(16,10, weight=values[47], color='blue')
N.add_edge(16,17, weight=values[48], color='blue')
N.add_edge(16,18, weight=values[49], color='blue')
N.add_edge(17,10, weight=values[50], color='blue')
N.add_edge(17,16, weight=values[51], color='blue')
N.add_edge(17,19, weight=values[52], color='blue')
N.add_edge(18,7, weight=values[53], color='blue')
N.add_edge(18,16, weight=values[54], color='blue')
N.add_edge(18,20, weight=values[55], color='blue')
N.add_edge(19,15, weight=values[56], color='blue')
N.add_edge(19,17, weight=values[57], color='blue')
N.add_edge(19,20, weight=values[58], color='blue')
N.add_edge(20,18, weight=values[59], color='blue')
N.add_edge(20,19, weight=values[60], color='blue')
N.add_edge(20,21, weight=values[61], color='blue')
N.add_edge(20,22, weight=values[62], color='blue')
N.add_edge(21,20, weight=values[63], color='blue')
N.add_edge(21,22, weight=values[64], color='blue')
N.add_edge(21,24, weight=values[65], color='blue')
N.add_edge(22,15, weight=values[66], color='blue')
N.add_edge(22,20, weight=values[67], color='blue')
N.add_edge(22,21, weight=values[68], color='blue')
N.add_edge(22,23, weight=values[69], color='blue')
N.add_edge(23,14, weight=values[70], color='blue')
N.add_edge(23,22, weight=values[71], color='blue')
N.add_edge(23,24, weight=values[72], color='blue')
N.add_edge(24,13, weight=values[73], color='blue')
N.add_edge(24,21, weight=values[74], color='blue')
N.add_edge(24,23, weight=values[75], color='blue')

# List of nodes to visit (including the start node)
delivery_nodes = [6, 11, 19, 24]
starting_node = 3
nodes_to_visit = [starting_node] + delivery_nodes

# Function to calculate the shortest paths between all relevant nodes
def compute_shortest_paths(graph, nodes):
    shortest_paths = {}
    for node1, node2 in itertools.combinations(nodes, 2):
        # Get shortest path from node1 to node2
        try:
            length = nx.shortest_path_length(graph, source=node1, target=node2, weight='weight')
            shortest_paths[(node1, node2)] = length
            shortest_paths[(node2, node1)] = length  # Ensure bidirectional paths
        except nx.NetworkXNoPath:
            shortest_paths[(node1, node2)] = float('inf')
            shortest_paths[(node2, node1)] = float('inf')
    return shortest_paths

# Function to calculate the total travel time of a specific route
def calculate_total_time(route, shortest_paths):
    total_time = 0
    for i in range(len(route) - 1):
        total_time += shortest_paths[(route[i], route[i + 1])]
    return total_time

# Compute all shortest paths between the nodes to visit
shortest_paths = compute_shortest_paths(N, nodes_to_visit)

# Generate all permutations of the delivery nodes (excluding the start node)
best_route = None
min_time = float('inf')

for perm in itertools.permutations(delivery_nodes):
    # Create a full route starting and ending at the starting node
    route = [starting_node] + list(perm) + [starting_node]

    # Calculate the total travel time for this route
    route_time = calculate_total_time(route, shortest_paths)

    # If this is the best route (minimizing travel time), update
    if route_time < min_time:
        min_time = route_time
        best_route = route

# Print the optimal route and its total travel time
print(f"Optimal Route: {best_route}")
print(f"Minimum Travel Time: {min_time} units")

# Time saved vs manually chosen route (example with a manual route)
manual_route = [starting_node, 6, 11, 24, 19, starting_node]
manual_time = calculate_total_time(manual_route, shortest_paths)

# Print the manually chosen route and its travel time
print(f"Manual Route: {manual_route}")
print(f"Manual Route Travel Time: {manual_time} units")

time_saved = manual_time - min_time
print(f"Time Saved (compared to manual route): {time_saved} units")
```

```python
import pandas as pd
import networkx as nx
import itertools

!pip install pulp

import pulp


file_path = '/content/drive/MyDrive/CEE201/pivoted_network_travel_times-Group_44.xlsx'
travel_times = pd.read_excel(file_path)

#capacity_file_path = '/content/drive/MyDrive/capacity_file-Group44.xlsx'
#capacity_values = pd.read_excel(capacity_file_path)

#file_path = '/content/drive/MyDrive/CEE201/pivoted_network_travel_times-Group_44.xlsx'
#travel_times = pd.read_csv(file_path)


values = []
for i in range(0, 76):
 values.append(travel_times.iloc[i, 1])

print(values[0])
N = nx.DiGraph()
N.add_nodes_from(range(1, 25))
N.add_edge(1, 2, weight=values[0], color='blue')
N.add_edge(1, 3, weight=values[1], color='blue')
N.add_edge(2, 1, weight=values[2], color='blue')
N.add_edge(2, 6, weight=values[3], color='blue')
N.add_edge(3, 1, weight=values[4], color='blue')
N.add_edge(3, 4, weight=values[5], color='blue')
N.add_edge(3, 12, weight=values[6], color='blue')
N.add_edge(4, 3, weight=values[7], color='blue')
N.add_edge(4, 5, weight=values[8], color='blue')
N.add_edge(4, 11, weight=values[9], color='blue')
N.add_edge(5, 4, weight=values[10], color='blue')
N.add_edge(5, 6, weight=values[11], color='blue')
N.add_edge(5, 9, weight=values[12], color='blue')
N.add_edge(6, 2, weight=values[13], color='blue')
N.add_edge(6, 5, weight=values[14], color='blue')
N.add_edge(6, 8, weight=values[15], color='blue')
N.add_edge(7, 8, weight=values[16], color='blue')
N.add_edge(7, 18, weight=values[17], color='blue')
N.add_edge(8, 6, weight=values[18], color='blue')
N.add_edge(8, 7, weight=values[19], color='blue')
N.add_edge(8, 9, weight=values[20], color='blue')
N.add_edge(8, 16, weight=values[21], color='blue')
N.add_edge(9, 5, weight=values[22], color='blue')
N.add_edge(9, 8, weight=values[23], color='blue')
N.add_edge(9, 10, weight=values[24], color='blue')
N.add_edge(10, 9, weight=values[25], color='blue')
N.add_edge(10, 11, weight=values[26], color='blue')
N.add_edge(10, 15, weight=values[27], color='blue')
N.add_edge(10, 16, weight=values[28], color='blue')
N.add_edge(10, 17, weight=values[29], color='blue')
N.add_edge(11, 4, weight=values[30], color='blue')
N.add_edge(11, 10, weight=values[31], color='blue')
N.add_edge(11, 12, weight=values[32], color='blue')
N.add_edge(11, 14, weight=values[33], color='blue')
N.add_edge(12, 3, weight=values[34], color='blue')
N.add_edge(12, 11, weight=values[35], color='blue')
N.add_edge(12, 13, weight=values[36], color='blue')
N.add_edge(13,12, weight=values[37], color='blue')
N.add_edge(13,24, weight=values[38], color='blue')
N.add_edge(14,11, weight=values[39], color='blue')
N.add_edge(14,15, weight=values[40], color='blue')
N.add_edge(14,23, weight=values[41], color='blue')
N.add_edge(15,10, weight=values[42], color='blue')
N.add_edge(15,14, weight=values[43], color='blue')
N.add_edge(15,19, weight=values[44], color='blue')
N.add_edge(15,22, weight=values[45], color='blue')
N.add_edge(16,8, weight=values[46], color='blue')
N.add_edge(16,10, weight=values[47], color='blue')
N.add_edge(16,17, weight=values[48], color='blue')
N.add_edge(16,18, weight=values[49], color='blue')
N.add_edge(17,10, weight=values[50], color='blue')
N.add_edge(17,16, weight=values[51], color='blue')
N.add_edge(17,19, weight=values[52], color='blue')
N.add_edge(18,7, weight=values[53], color='blue')
N.add_edge(18,16, weight=values[54], color='blue')
N.add_edge(18,20, weight=values[55], color='blue')
N.add_edge(19,15, weight=values[56], color='blue')
N.add_edge(19,17, weight=values[57], color='blue')
N.add_edge(19,20, weight=values[58], color='blue')
N.add_edge(20,18, weight=values[59], color='blue')
N.add_edge(20,19, weight=values[60], color='blue')
N.add_edge(20,21, weight=values[61], color='blue')
N.add_edge(20,22, weight=values[62], color='blue')
N.add_edge(21,20, weight=values[63], color='blue')
N.add_edge(21,22, weight=values[64], color='blue')
N.add_edge(21,24, weight=values[65], color='blue')
N.add_edge(22,15, weight=values[66], color='blue')
N.add_edge(22,20, weight=values[67], color='blue')
N.add_edge(22,21, weight=values[68], color='blue')
N.add_edge(22,23, weight=values[69], color='blue')
N.add_edge(23,14, weight=values[70], color='blue')
N.add_edge(23,22, weight=values[71], color='blue')
N.add_edge(23,24, weight=values[72], color='blue')
N.add_edge(24,13, weight=values[73], color='blue')
N.add_edge(24,21, weight=values[74], color='blue')
N.add_edge(24,23, weight=values[75], color='blue')






# Define parameters
source_node = 2  # I1 (Industrial Hub)
destination_node = 23  # P3 (Coastal Port)
truck_demand = 500  # Number of trucks per week

# Initialize the optimization model
model = pulp.LpProblem("Minimize_Travel_Time_and_Emissions", pulp.LpMinimize)

# Define decision variables for the flow of trucks on each edge
flow_vars = {}
for u, v, data in N.edges(data=True):
    flow_vars[(u, v)] = pulp.LpVariable(f"flow_{u}_{v}", lowBound=0, cat="Continuous")

# Define the objective function
# Total travel time: sum of (travel time * flow) for all edges
# Total emissions: sum of (emissions * flow) for all edges
travel_time = pulp.lpSum(data['weight'] * flow_vars[(u, v)] for u, v, data in N.edges(data=True))
emissions = pulp.lpSum((5 if data['weight'] < 20 else 0.25 * data['weight']) * flow_vars[(u, v)] for u, v, data in N.edges(data=True))

# Trade-off parameter (alpha) to balance time and emissions
alpha = 0.5  # Adjust between 0 and 1 for different trade-offs
model += alpha * travel_time + (1 - alpha) * emissions

# Flow conservation constraints
for node in N.nodes:
    if node == source_node:
        model += pulp.lpSum(flow_vars[(u, v)] for u, v in N.out_edges(node)) - \
                 pulp.lpSum(flow_vars[(v, u)] for v, u in N.in_edges(node)) == truck_demand
    elif node == destination_node:
        model += pulp.lpSum(flow_vars[(u, v)] for u, v in N.out_edges(node)) - \
                 pulp.lpSum(flow_vars[(v, u)] for v, u in N.in_edges(node)) == -truck_demand
    else:
        model += pulp.lpSum(flow_vars[(u, v)] for u, v in N.out_edges(node)) - \
                 pulp.lpSum(flow_vars[(v, u)] for v, u in N.in_edges(node)) == 0

# Solve the model
solver = pulp.PULP_CBC_CMD(msg=True)
model.solve(solver)

# Print the solution status
print(f"Solution Status: {pulp.LpStatus[model.status]}")

# Print the optimal values for flow variables
print("\nOptimal Truck Flows on Edges:")
for u, v in N.edges:
    flow_value = flow_vars[(u, v)].varValue
    if flow_value > 0:
        print(f"Edge ({u}, {v}): {flow_value:.2f} trucks/week")

# Calculate total travel time and emissions from the solution
total_travel_time = sum(data['weight'] * flow_vars[(u, v)].varValue for u, v, data in N.edges(data=True))
total_emissions = sum((5 if data['weight'] < 20 else 0.25 * data['weight']) * flow_vars[(u, v)].varValue for u, v, data in N.edges(data=True))

# Print the summary
print("\nSummary of the Solution:")
print(f"Total Travel Time: {total_travel_time:.2f} vehicle-hours/week")
print(f"Total CO2 Emissions: {total_emissions:.2f} units/week")
```
